```python
import numpy as np 
import matplotlib.pyplot as plt
import glob
import os
#import csv
import pandas as pd
import matplotlib 
import matplotlib.pylab as pylab
from mpl_toolkits.basemap import Basemap
pylab.rcParams['figure.figsize'] = 12, 8
pd.options.display.mpl_style = 'default'
%matplotlib inline

# read in all files in current directory, assuming that they have the file struction nt.csv or nh.csv, eg 1t.csv, 13t.csv, 40h.csv, etc 
files = glob.glob('*[tT].csv') # read in data files 
meta = pd.DataFrame(pd.read_csv('bmoremetadata.csv', sep = ',')) # read in metadata 
meta = meta.set_index(meta['sensornumber']) # set the row names for metadata to be the sensor number

for index in meta.index: # loop over the sensor numbers in metadata and check if they are numbers
    try: 
        if str.isdigit(index)== 0: 
            meta = meta.drop(meta.loc[index]) # delete the rows without sensor numbers
    except (TypeError):
        pass
        
meta = meta.sort(axis=0) # sort the data in order of the sensor number 
meta = meta.set_index(meta['sensornumber'].astype(int))
# read in all the csv files into a Panda's dataframe, skipping the headers of the csv files
frames = []
sensornumbers = []
for file in files: 
    try : 
        frames.append(pd.read_csv(file, sep = ',', skiprows = 19))
        sensornumbers.append(int(os.path.splitext(file)[0][0:-1]) )
    except ValueError: 
        print "oops... something went wrong"
    
data = pd.concat(frames, axis =1)

# Clean up data and build a new dataset
# next, normalize the dates: the sensors probably weren't set to all go off at the same time
# assume they all are within an hour of each other
maxStart = data['Date/Time'].values[0,:].max() # the start time of the last sensor to start
minEndIndex = pd.isnull(data['Date/Time']).any(1).nonzero()[0].min() # find the earliest row instance of Nans
maxEnd = data['Date/Time'].values[minEndIndex-1,:].max() #picks the latest time in that row
rng = pd.date_range(maxStart, maxEnd, freq='H') # make a date range from maxStart to maxEnd, with hourly frequency

Temp = data['Value'].values[0:minEndIndex,:] #The temperature data 
clim = Temp.mean(axis=1) # the temperature 'climatology'
anomaly = Temp - np.tile(clim[:,np.newaxis], (1, Temp.shape[1])) # the anomaly data

# produces a data structure with time from variable rng as the rows and sensornumbers as the column heading
# the column 
tempDF = pd.DataFrame(data['Value'].values[0:minEndIndex,:], rng[:-1], sensornumbers).sort(axis=1)
anomalyDF = pd.DataFrame(anomaly, rng[:-1], sensornumbers).sort(axis=1)

#clean the metadata so that missing sensors don't mess with analysis
meta = meta.loc[np.intersect1d(tempDF.columns.values, meta.sensornumber.values)]
```

```python
#U = tempDF[eagerSensors.astype(int)].mean(axis=0).values
#X = meta['location:Longitude'].values[eager] 
#Y = meta['location:Latitude'].values[eager] 

# The semi-variogram is given by 
# SV(h) = 1/ (2N(h)) * \sum_{N(h)}(z_i-z_j)^2 
# see more at: http://connor-johnson.com/2014/03/20/simple-kriging-in-python/
# First, compute the squared diferences and the distances h 
# Next, bin them in order to take the average

# For this function, however, we'll just do differences vs. distance

# X,Y are vectors for longitude and latitude 
# Y is the data vector, same length as X or Y
def variogram(X,Y,U) :
    d = np.zeros([U.size, U.size]) 
    v = np.zeros([U.size, U.size]) 
    # compute the differences and distnaces
    for i in range(0,U.size) : 
        for j in range(0, U.size) : 
            if i >j : 
                d[i,j] = ((X[i]-X[j])**2 + (Y[i]-Y[j])**2)**.5
                v[i,j] = np.abs(U[i]-U[j])
    d = d*2*np.pi*(6.371e3)/360. # change distance from lat/lon to km 
    #n, bins, patches = plt.hist(d[d>0]) 
    hist, bins = np.histogram(d[d>0], bins = 20)
    vline = np.zeros(bins.shape[0]-1) 
    for i in range(0, bins.shape[0]-1) : 
        a,b = np.where( (bins[i] <= d) & (d < bins[i+1])) 
        vline[i] = np.mean(v[a,b])
    return d, v, vline, bins
#d,v,vline, bins = variogram(X,Y,U)
```

```python
# Plot distance versus temperature difference along each street 
i = 0
plt.figure(figsize=(12, 20))
for sensor in allSensors: 
    #X = meta['location:Latitude'].values[ allSensors[sensor][0]]
    #U = tempDF[allSensors[sensor][1].astype(int)].mean(axis=0).values
    d, v, vline, bins= variogram(meta['location:Longitude'].values[ allSensors[sensor][0]],
                                 meta['location:Latitude'].values[ allSensors[sensor][0]],
                                 tempDF[allSensors[sensor][1].astype(int)].mean(axis=0).values)
    plt.subplot(len(allSensors), 1,i)
    plt.scatter(d[d>0], v[d>0])
    plt.plot(bins[0:-1], vline, '-or') 
    plt.hold(True)
    #plt.plot(bins[0:-1], variogram)

    plt.title('Distances vs. temp. difference along %s Street'%sensor)
    plt.xlabel('Distance in km')
    plt.ylabel('Temperature')
    i = i+1
```

```python
# Plot distance versus temperature difference along the NS streets, aggregated 

NSsensors = {'NS': [np.concatenate((wash, mont), axis =1), np.concatenate((washSensors, montSensors)) ]}

i = 0
plt.figure(figsize=(12, 10))
for sensor,item in NSsensors.iteritems(): 
    #X = meta['location:Latitude'].values[ allSensors[sensor][0]]
    #U = tempDF[allSensors[sensor][1].astype(int)].mean(axis=0).values
    d, v, vline, bins= variogram(meta['location:Latitude'].values[ item[0]].T,
                                 meta['location:Longitude'].values[ item[0]].T,
                       tempDF[NSsensors[sensor][1].astype(int)].mean(axis=0).values)
    plt.subplot(len(NSsensors), 1,i)
    plt.scatter(d[d>0], v[d>0])
    plt.plot(bins[0:-1], vline, '-or') 
    plt.hold(True)
    #plt.plot(bins[0:-1], variogram)

    plt.title('Distances vs. temp. difference along %s Streets'%sensor)
    plt.xlabel('Distance in km')
    plt.ylabel('Temperature')
    i = i+1
```
