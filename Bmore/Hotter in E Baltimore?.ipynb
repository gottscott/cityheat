```python
# Import data and libraries
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
import pandas as pd
import matplotlib
import matplotlib.pylab as pylab
from mpl_toolkits.basemap import Basemap
pylab.rcParams['figure.figsize'] = 16, 12
pd.options.display.mpl_style = 'default'
get_ipython().magic(u'matplotlib inline')
import ibuttonplots
import importibuttondata
%matplotlib inline

files = glob.glob('./data/*[tT].csv')
meta0 = pd.DataFrame(pd.read_csv('./data/bmoremetadata.csv', sep = ','))
tempDF0, anomalyDF0, meta0 = importibuttondata.importdata(files, meta0)

#files = glob.glob('./data/*[tT].csv')
files = glob.glob('./data/falldownload/*[tT].csv')
meta1 = pd.DataFrame(pd.read_csv('./data/falldownload/TempSensorFinal_results-4.csv', sep = ','))
#meta0 = pd.DataFrame(pd.read_csv('./data/bmoremetadata.csv', sep = ','))
tempDF1, anomalyDF1, meta1 = importibuttondata.importdata(files, meta1)

frames = [tempDF0, tempDF1]
tempDF = pd.concat(frames)
anomalyDF = pd.concat([anomalyDF0, anomalyDF1])
meta = pd.concat([meta0, meta1.loc()[np.setdiff1d(meta1.sensornumber, meta0.sensornumber)]])
```

```python
# Import Science Center observation data
date_spec = {'Date': [2, 3]}
urbandf = pd.read_csv('./data/observations/urban/2138216847366dat.csv', 
                      sep = ',', 
                      parse_dates = {'Date':[2, 3]},
                      #parse_dates = ['YYYYMMDD', 'HHM'],
                      keep_date_col=True).set_index('Date')
#urbandf.Temp[urbandf.Temp == '****'] = 'NaN' # replace stars with nans
urbandf.Temp = (urbandf.Temp.astype('float')) #-32)*5/9. #change the data to float
urbandf.Temp[urbandf.Temp > 400] = 'NaN' #change unrealistic temp data to nan, data in F so 400 too high
urbandf = urbandf[~((urbandf.index.month == 2) & (urbandf.index.day == 29))] # get rid of leap years
#df.DEWP[df.DEWP == '****'] = 'NaN' # replace stars with nans
#df.DEWP = df.DEWP.astype('float') #change the data to float
urbandf = urbandf.resample('H')
urbandf.rename(columns = {'Temp' : 'TEMP'}, inplace = True)
urbandf.index = urbandf.index - pd.DateOffset(hours  = 4)
```

```python
ebaltsensorsi = np.where(meta['location:Longitude']>= -76.61)# -76.6072591 )
plt.scatter(meta['location:Longitude'].iloc[ebaltsensorsi],
            meta['location:Latitude'].iloc[ebaltsensorsi])
ebaltsensors = meta.sensornumber.iloc[ebaltsensorsi]
selected = ebaltsensors
```

```python
# Is it hotter in E. Baltimore than downtown at the science center? 
# Mean differences 
#
# Mean temperature differences between E. Balt and science center 
start = '2015-08-20'
end = '2015-9-10'
time = tempDF[start:end].index
diff = tempDF[start:end][selected]-urbandf[start:end].TEMP
buttontemp = tempDF[start:end][selected].mean(axis=1)
obs =urbandf[start:end].TEMP
# Time Series of E. Balt mean temperature and Science Center 
fig = plt.figure(figsize=[10,5])
plt.plot(time, buttontemp, 'o-k', label = 'iButton Mean', linewidth = 3)
obs.plot(label = 'Sci Center', linewidth = 3, style = 'o-')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., ncol = 1)
plt.title('Temperature')
plt.ylabel('Temperature in $^\circ $C')

# Time series of difference
fig = plt.figure(figsize=[10,5])
plt.plot(time, diff)
lgd = plt.legend(selected, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., ncol = 6)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., ncol = 1)
plt.title('E. Baltimore Average - Science Center')
plt.ylabel('Temperature in $^\circ $C')
plt.xticks(rotation=30)
plt.show()

# Scatter plot of difference vs temp 
fig = plt.figure(figsize=[10,5])
x = obs
y = diff.mean(axis=1)
plt.scatter(x,y, 
            label = '$\Delta T$')

m,b = np.polyfit(x, y,1)
plt.plot(x, m*x+b, 
             color = pd.tools.plotting._get_standard_colors(3)[0],
             lw =4,
             label = '$\Delta T=%.2f \cdot T_{BWI}+%.2f$'%(m,b),
             )
plt.title('Buttons - Science Center vs Temperature ')
plt.xlabel('Temperature from Science Center Station')
plt.ylabel('$\Delta T$ (Buttons- Science Center)')
plt.xticks(rotation=30)
lgd = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., ncol = 1)
plt.show()

# Histograms of Differences
fig = plt.figure(figsize=[10,5])
data = diff.mean(axis=1)
data = data[~np.isnan(data)]
n,bins, patches= plt.hist(data,20)
plt.axvline(0, color = 'k') # plot zero line
plt.axvline(data.mean(),
            linestyle='dashed',
            color = pd.tools.plotting._get_standard_colors(3)[2],
            linewidth=2,
            label = 'mean, %2.1f $^\circ$ C'%np.nanmean(data),
            alpha = 1.0)

plt.axvline(data.std()+data.mean(),
            color = pd.tools.plotting._get_standard_colors(3)[1],
            linestyle='dashed',
            linewidth=2,
            label = '$\pm \sigma$, %2.1f $^\circ$ C'%np.nanstd(data)
            )

plt.axvline(data.mean()-data.std(),
            color = pd.tools.plotting._get_standard_colors(3)[1],
            linestyle='dashed',
            linewidth=2,
            )
plt.xlim([-5,5])
plt.xlabel('Temperature in $^\circ$ C')
plt.ylabel('Count')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #, ncol = 3)
plt.title('Histogram of Buttons - Science Center')

# Diurnal Cycle
fig = plt.figure(figsize=[10,5])
buttontemp.groupby(buttontemp.index.hour).mean().plot(
    #yerr = ruraldf['2015-06-25':'2015-08-09'].TEMP.groupby(ruraldf['2015-06-25':'2015-08-09'].index.hour).std(),
    label = 'iButtons',
    style = '-ok',
    linewidth = 3,
    #color = pd.tools.plotting._get_standard_colors(4)[2]
    )
maximum =buttontemp.groupby(buttontemp.index.hour).mean().max()
plt.text(15.5, maximum,'%2.1f'%maximum,
                 bbox=dict(facecolor='white', edgecolor = '#636363')
                 )
minimum = buttontemp.groupby(buttontemp.index.hour).mean().min()
plt.text(5.5, minimum,'%2.1f'%minimum,
                 bbox=dict(facecolor='white', edgecolor = '#636363')
                 )

obs.groupby(obs.index.hour).mean().plot(
    label = 'Science Center', 
    #yerr = urbandf['2015-06-25':'2015-08-09'].TEMP.groupby(urbandf['2015-06-25':'2015-08-09'].index.hour).std(),
    linewidth = 3,
    style = '-o',
    #color = pd.tools.plotting._get_standard_colors(4)[2],  
   )

maximum =obs.groupby(obs.index.hour).mean().max()
plt.text(15.5, maximum,'%2.1f'%maximum,
                 bbox=dict(facecolor='white', edgecolor = '#636363')
                 )
minimum = obs.groupby(obs.index.hour).mean().min()
plt.text(5.5, minimum,'%2.1f'%minimum,
                 bbox=dict(facecolor='white', edgecolor = '#636363')
                 )

lgd = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., ncol = 1)
plt.xlabel('Hour')
plt.ylabel('Temp')
plt.title('Diurnal Cycle')

```
