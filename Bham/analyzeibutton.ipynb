```python
# Import data and libraries
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
import matplotlib
%matplotlib inline
import pandas as pd
import matplotlib.pylab as pylab
from mpl_toolkits.basemap import Basemap
pylab.rcParams['figure.figsize'] = 16, 12
pd.options.display.mpl_style = 'default'
get_ipython().magic(u'matplotlib inline')
import sys 

sys.path.append('../Bmore/')
import ibuttonplots
import importibuttondata

files = glob.glob('./data/*[tT].csv')
#files = glob.glob('./data/olddata/*[tT].csv')
meta0 = pd.DataFrame(pd.read_csv('./data/birminghammeta_cleaned.csv', sep = ','))
tempDF, anomalyDF, meta = importibuttondata.importdata(files, meta0)
```

```python
files = glob.glob('./data/stationdata/*.csv')
date_spec = {'Date': [ 2]}
frames = []
i = 0
for file in files:
    data = pd.read_csv(file, sep = ',', 
                              index_col=0,
                              parse_dates = date_spec, 
                              keep_date_col=True)
    data.TMIN = data.TMIN*.1
    data.TMAX = data.TMAX*.1
    data1 = data.copy()
    stations = set(data.STATION)
    #frames = []

    for station in stations: 
        tmax =  data[data.STATION==station].TMAX
        tmin = data1[data1.STATION==station].TMIN
        frames.append( tmax.combine_first(tmin))
        frames[i].name = station 
        i = i+1

        #    frames.append({station : tmax.combine_first(tmin)})

obsDF = pd.concat(frames, axis =1).dropna()
obsDF
```
